As 3 primeiras linhas pertencem ao cluster 0 enquanto as Ãºltimas duas pertencem ao cluster 1.

```python
X, _ = make_blobs( n_samples=300,
                   centers=4,
                   cluster_std=0.6,
                   random_state=0 )
```

```python
X[:10]
```

```python
plt.scatter(X[:,0], X[:,1]);
```

```python
# model definition
model = AffinityPropagation()

# model training
model.fit( X )

# model fit (clustering)
labels = model.predict( X )
```

```python
fig = plt.figure()

plt.title( 'Clustering' )
plt.scatter( X[:, 0], X[:, 1], c=labels, alpha=0.7 );
```

```python
print( f'NÃºmero de Clusters = {len(np.unique(labels))}' )
```

```python
# model definition
model = AffinityPropagation( preference=-50 )

# model training
model.fit( X )

# model fit (clustering)
labels = model.predict( X )
```

```python
fig = plt.figure()

plt.title( 'Clustering' )
plt.scatter( X[:, 0], X[:, 1], c=labels, alpha=0.7 );
```

```python
print( f'NÃºmero de Clusters = {len(np.unique(labels))}' )
```

```python
p = np.arange( -1, -60, -1 )

fig = plt.figure()

for i in p:
    # model definition
    model = AffinityPropagation( preference=i )

    # model training
    model.fit( X )

    # model fit (clustering)
    labels = model.predict( X )


    # show clusters
    plt.scatter( X[:, 0], X[:, 1], c=labels, alpha=0.7 )

    display.clear_output( wait=True )
    plt.show()
    plt.gcf()
    
    time.sleep( 0.05 )
```
